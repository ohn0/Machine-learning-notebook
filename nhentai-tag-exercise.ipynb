```python
import matplotlib.pyplot as plt
def plotScatterColorMap(x, y, plotter):
    plotter.plot(kind="scatter", x=x, y=y, alpha=0.2,figsize=(10,7), cmap=plt.get_cmap("jet"), colorbar=True, s=plotter[y])
```

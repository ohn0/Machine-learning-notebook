```python
import os
import pandas as pd

NYC_RATS_PATH = "C:\\Users\\neel\\jupyter\\datasets\\NycRats\\Rat_Sightings.csv"

def load_csv_into_frame(csv_path):
    return pd.read_csv(csv_path)
    
```

```python
rats = load_csv_into_frame(NYC_RATS_PATH)

# print(rats["Longitude"].min())
# print(rats["Longitude"].max())
# print(rats["Latitude"].min())
# print(rats["Latitude"].max())
```

```python
z = rats["Street Name"].value_counts()
```

```python
from sklearn.preprocessing import OrdinalEncoder

ordinal_encoder = OrdinalEncoder()
borough_encoded = ordinal_encoder.fit_transform(rats[["Borough"]])
borough_encoded[:10]
ordinal_encoder.categories_

rats["Borough_encoded"] = borough_encoded
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
rats.plot(kind="scatter", x="Longitude", y="Latitude",s=z[rats["Street Name"]]/300, alpha=0.5, figsize=(20,17),c=z[rats["Street Name"]], cmap=plt.get_cmap("jet"), )
# plt.legend()
```
